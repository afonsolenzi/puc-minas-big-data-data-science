```python
client = MongoClient()
db = client['teste']
#select the collection within the database
test = db.truck_event_text_partition
#convert entire collection to Pandas dataframe
df_truck_event = pd.DataFrame(list(test.find()))
```

```python
merged_inner_overspeed = merged_inner[merged_inner['eventType']=='Overspeed']
```
